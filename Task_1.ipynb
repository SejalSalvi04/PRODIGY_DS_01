### Importing necessary libraries 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

### Loading the dataset

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv",
    skiprows=4)

df.head()
```

### Data info

```python
df.info()
```

### Check the null values

```python
df.isnull().sum()
```

### Bar Chart 

```python
latest_year = '2024'

top10 = (
    df[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(10)
)
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=top10, x='Country Name', y=latest_year,palette="inferno")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Population (2023)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Histogram  : Distribution of Population Across Coutries

```python
plt.figure(figsize=(7,5))
sns.histplot(data=df, x=latest_year,kde=True, bins=15)
plt.xlabel("Population")
plt.ylabel("No. of countries")
plt.grid()
plt.title("Population Distribution (2023)")
plt.show()
```
